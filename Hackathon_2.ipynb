# Oppgave:
<p style="font-size: 16px; font-family: 'Arial', sans-serif; color: #2c3e50;">I morgen tidlig, planlegger jeg å snike meg ut av seminaret og flyr direkte enten til London eller Bucuresti. Jeg vil i begge tilfeller være der ca. kl. 12, men hvor jeg drar avhenger av været.</p> 
<ul style="font-size: 16px; font-family: 'Arial', sans-serif; color: #2c3e50;">
    <li><b>Hvor kommer været til å være best - London eller Bucuresti?</b> Oppgi begrunnelse og urler.</li>
</ul>

```python
url = "https://api.met.no/weatherapi/locationforecast/2.0/compact?lat=59.91&lon=10.76"

headers = {'User-Agent': 'API-opplæring arne.martin.fevolden@udir.no', 
            'Accept-Encoding': 'gzip, deflate, br', 
            'Accept': '*/*', 
            'Connection': 'keep-alive'}

# API spørring
response = requests.get(url, headers=headers)
data = response.json()
print(data)
```

```python
yr_data_list = []

for i in response.json()["properties"]['timeseries']:
    try:
        yr_data_list.append((
            i['time'], 
            i['data']['instant']['details']['cloud_area_fraction'],  # Instant cloud coverage
            i['data']['instant']['details']['air_temperature'],  # Instant temperature
            i['data']['instant']['details']['relative_humidity'],  # Instant relative humidity
            i['data']['instant']['details']['wind_speed'],  # Instant wind speed
            i['data']['next_1_hours']['summary']['symbol_code']  # Forecast symbol code for next 1 hour
        ))
    except KeyError:
        continue  # Skip entries where the required data is missing

df = pd.DataFrame(yr_data_list, columns=[
    'time', 'cloud_area_fraction', 'air_temperature', 'relative_humidity', 'wind_speed', 'symbol_code_next_1_hour'
])

df
```
